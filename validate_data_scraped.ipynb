```python
import pandas as pd 
```

```python

# =============================================
INPUT_FILE = "./PART10/PC10_(213308-237007).csv"
OUTPUT_FILE = "./PART10/Output_PC10_19Jan.csv"
# =============================================
# LOAD DATA
# =============================================

# =============================================
# Load input
input_df = pd.read_csv(INPUT_FILE, encoding="utf-8-sig", sep=None, engine="python")
input_df.columns = ['idsbr', 'Query']
input_df['idsbr'] = input_df['idsbr'].astype(str).str.replace('Ã¯Â»Â¿', '', regex=False).str.strip()
# Load output
output_df = pd.read_csv(OUTPUT_FILE)
output_df['idsbr'] = output_df['idsbr'].astype(str).str.strip()
# =============================================
# PERBANDINGAN
# =============================================
input_idsbr = set(input_df['idsbr'])
output_idsbr = set(output_df['idsbr'])
missing = input_idsbr - output_idsbr  # Ada di input tapi tidak ada di output
extra = output_idsbr - input_idsbr    # Ada di output tapi tidak ada di input
print("=" * 50)
print("ðŸ“Š PERBANDINGAN INPUT vs OUTPUT")
print("=" * 50)
print(f"Total idsbr di INPUT       : {len(input_idsbr)}")
print(f"Total unique idsbr di OUTPUT: {len(output_idsbr)}")
print(f"Total rows di OUTPUT        : {len(output_df)}")
print("-" * 50)
print(f"âŒ idsbr BELUM di-scrape    : {len(missing)}")
print(f"âš ï¸  idsbr EXTRA di output   : {len(extra)}")
print("=" * 50)
# =============================================
# DETAIL STATUS DI OUTPUT
# =============================================
print("\nðŸ“ˆ BREAKDOWN STATUS:")
print(output_df['Status'].value_counts())
# =============================================
# TAMPILKAN YANG BELUM DI-SCRAPE (jika ada)
# =============================================
if missing:
    print(f"\nâŒ Daftar idsbr yang BELUM di-scrape ({len(missing)}):")
    for m in list(missing)[:20]:  # Tampilkan max 20
        print(f"  - {m}")
    if len(missing) > 20:
        print(f"  ... dan {len(missing) - 20} lainnya")
else:
    print("\nâœ… SEMUA idsbr dari INPUT sudah ada di OUTPUT!")
# =============================================
# CEK YANG ERROR
# =============================================
error_df = output_df[output_df['Status'] == 'Error']
if len(error_df) > 0:
    error_idsbr = error_df['idsbr'].unique()
    print(f"\nâš ï¸  idsbr dengan Status ERROR ({len(error_idsbr)}):")
    for e in error_idsbr[:10]:  # Tampilkan max 10
        print(f"  - {e}")
    if len(error_idsbr) > 10:
        print(f"  ... dan {len(error_idsbr) - 10} lainnya")
```
